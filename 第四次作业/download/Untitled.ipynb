```python
import pandas as pd
import numpy as np
import os
```

```python
# 使用列表推导式
csv_files = [file for file in os.listdir(".") if file.endswith(".csv")]
print("当前文件夹下的CSV文件：")
for file in csv_files:
    print(file)
```

```python
# Sub_List = ['Chemistry','Biology&Biochemistry','MaterialsScience','EnvironmentEcology','Engineering','PHYSICS','SOCIALSCIENCES','COMPUTER-SCIENCE','PLANT-ANIMAL-SCIENCE','MOLECULAR-BIOLOGY&GENETICS',
#         'CLINICAL-MEDICINE','PSYCHIATRY-PSYCHOLOGY','NEUROSCIENCE&BEHAVIOR','MATHEMATICS','AGRICULTURAL-SCIENCES','PHARMACOLOGY&TOXICOLOGY']

for csv_file in csv_files:
    Sub = csv_file[:-4]
    print(Sub)
    # 学科列表，每个列表项用于添加到url中
    df = pd.read_csv(csv_file, encoding='latin1', skiprows=1)
    df = df.drop(df.index[len(df) - 1])
    df.tail(1)
    df['Rank'] = df['Unnamed: 0'].astype(int)
    Total_Insstitues = len(df)
    if np.sum(df['Institutions']=="EAST CHINA NORMAL UNIVERSITY") == 0:
        continue
    Current_rank = df[df['Institutions']=="EAST CHINA NORMAL UNIVERSITY"]['Rank'].values[0]
    Rank_Percent = Current_rank / Total_Insstitues
    print("{0}的排名为{1}%".format(Sub, Rank_Percent))
    print("华师大的具体数据如下：")
    print(df[df['Institutions']=="EAST CHINA NORMAL UNIVERSITY"])
    P0001 = int(np.ceil(Total_Insstitues*0.01))
    P001 = int(np.ceil(Total_Insstitues*0.1))
    print("{}的前0.01%高校为".format(Sub))
    print(df[df['Rank']==P0001])
    print("{}的前0.1%高校为".format(Sub))
    print(df[df['Rank']==P001])
```
